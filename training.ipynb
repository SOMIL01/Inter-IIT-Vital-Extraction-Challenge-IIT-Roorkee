```python
import os
for (dirname, dirs, files) in os.walk('/content/content/yolov5/runs/detect/exp5/labels'):
  for filename in files:
    filepath = os.path.join('/content/content/yolov5/runs/detect/exp5/labels',filename) 
    with open(filepath,'r+') as f:
        lines = f.readlines()
        count1 = len(lines)
        arr = [[],[],[],[],[],[]]
        str_arr = ['','','','','','']
        count = 0
        for line in lines:
            count += 1
            str1 = line.split()
            str2 = str1
            str2.pop(-1)
            line1 = ' '.join(str2)
            line1 += '\n'
            for i in range(0,6):
                if(line[0] == str(i)):
                    if(len(arr[i]) == 0):
                        arr[i] = str1
                        str_arr[i] = line1
                    else:
                        if(float(arr[i][-1]) < float(str1[-1])):
                            arr[i] = str1
                            str_arr[i] = line1
        f.close()
    with open(filepath,'w') as f:
        f.seek(0)                   
        for i in range(0,count1):
            f.write('')
        f.seek(0)
        for i in str_arr:
            if(i != ''):
                f.write(i)
```

```python
from random import choice
#import shutil

#arrays to store file names
imgs =[]
xmls =[]

#setup dir names
trainPath = '/content/dataset/images/train'
valPath = '/content/dataset/images/val'
crsPathimg = '/content/images' #dir where images and annotations stored
crspathlbl= '/content/data'

#setup ratio (val ratio = rest of the files in origin dir after splitting into train and test)
train_ratio = 0.95
val_ratio = 0.05


#total count of imgs
totalImgCount = 2911

#soring files to corresponding arrays
for (dirname, dirs, files) in os.walk(crsPathimg):
    for filename in files:
           imgs.append(filename)

for (dirname, dirs, files) in os.walk(crspathlbl):
    for filename in files:
            xmls.append(filename)
         


#counting range for cycles
countForTrain = int(len(imgs)*train_ratio)
countForVal = int(len(imgs)*val_ratio)
print("training images are : ",countForTrain)
print("Validation images are : ",countForVal)
```

```python
trainimagePath = '/content/dataset/images/train'
trainlabelPath = '/content/dataset/labels/train'
valimagePath = '/content/dataset/images/val'
vallabelPath = '/content/dataset/labels/val'
#cycle for train dir
for x in range(countForTrain):

    fileJpg = choice(imgs) # get name of random image from origin dir
    fileXml = fileJpg[:-4] +'txt' # get name of corresponding annotation file

    #move both files into train dir
    #shutil.move(os.path.join(crsPath, fileJpg), os.path.join(trainimagePath, fileJpg))
    #shutil.move(os.path.join(crsPath, fileXml), os.path.join(trainlabelPath, fileXml))
    shutil.copy(os.path.join(crsPathimg, fileJpg), os.path.join(trainimagePath, fileJpg))
    shutil.copy(os.path.join(crspathlbl, fileXml), os.path.join(trainlabelPath, fileXml))


    #remove files from arrays
    imgs.remove(fileJpg)
    xmls.remove(fileXml)



#cycle for test dir   
for x in range(countForVal):

    fileJpg = choice(imgs) # get name of random image from origin dir
    fileXml = fileJpg[:-4] +'txt' # get name of corresponding annotation file

    #move both files into train dir
    #shutil.move(os.path.join(crsPath, fileJpg), os.path.join(valimagePath, fileJpg))
    #shutil.move(os.path.join(crsPath, fileXml), os.path.join(vallabelPath, fileXml))
    shutil.copy(os.path.join(crsPathimg, fileJpg), os.path.join(valimagePath, fileJpg))
    shutil.copy(os.path.join(crspathlbl, fileXml), os.path.join(vallabelPath, fileXml))
    
    #remove files from arrays
    imgs.remove(fileJpg)
    xmls.remove(fileXml)

#rest of files will be validation files, so rename origin dir to val dir
#os.rename(crsPath, valPath)
#shutil.move(crsPath, valPath)
```
